```python
import pandas as pd

# Load the Excel file
file_path = r'D:\EEX_Auction_Calendar_21.06.2023.xlsx' # Update with your file path
df = pd.read_excel(file_path, sheet_name="Details_EUA")

# Create an empty DataFrame for the wide format data
wide_df = pd.DataFrame()

# Define the number of rows in each chunk
chunk_size = 42

# Iterate through the chunks and append them side by side
for i in range(0, len(df), chunk_size):
    chunk = df.iloc[i:i+chunk_size]  # Select the chunk
    chunk.columns = range(len(chunk.columns))  # Reset column names to integers
    wide_df = pd.concat([wide_df, chunk], axis=1)

# Reset column names in the wide format DataFrame
wide_df.columns = range(len(wide_df.columns))

# Save the wide format DataFrame to a new Excel file
wide_file_path = "wide_format_data.xlsx"  # Update with your desired file path
wide_df.to_excel(wide_file_path, index=False)

print("Data has been transformed and saved to", wide_file_path)
```

```python
import pandas as pd

# Load the Excel file
excel_file = pd.ExcelFile(r'D:\EEX_Auction_Calendar_21.06.2023.xlsx')

# Name of the sheet with the tables
sheet_name = "Details_EUA"

# Fixed length of each table
table_length = 42

# Initialize an empty list to store the tables
all_tables = []

# Loop through the rows, skipping rows in between the tables
for start_row in range(0, len(excel_file.parse(sheet_name)) - table_length + 1, table_length):
    # Read each table into a DataFrame
    table_data = excel_file.parse(sheet_name, skiprows=start_row, nrows=table_length)
    all_tables.append(table_data)

# Concatenate the tables horizontally while preserving the first column in the first table
merged_tables = [all_tables[0]]
for table in all_tables[1:]:
    merged_tables.append(table.iloc[:, 1:])  # Exclude the first column in subsequent tables

# Concatenate all the tables horizontally
result = pd.concat(merged_tables, axis=1)

# Print the merged DataFrame
print(result)
```
