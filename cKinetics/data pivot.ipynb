```python
import pandas as pd
from io import StringIO
```

```python
df = pd.read_csv(r'D:\vedansh.csv', encoding='utf-8')

df.head()
```

```python
df_pivoted = df.pivot_table(index=df.groupby('column_names').cumcount(), columns='column_names', values='column_values', aggfunc='first')
```

```python
df_pivoted.to_csv(r'D:\vedansh_cleaned.csv',encoding='utf-8', index=False, errors='replace')
```

```python
df = pd.read_excel(r'D:\testv.xlsx')
df.head()
```

```python
df['index1'].fillna(method='ffill', inplace=True)
```

```python
df
```

```python
groups = (df['index2'] == 1).cumsum()

# Calculate the incrementing sequence within each group
incrementing_sequence = df.groupby(groups).cumcount() + 1

# Fill the 'index2' column with the incrementing sequence
df['index2'] = df['index2'].fillna(incrementing_sequence)
```

```python
df.to_csv(r'D:\testv-cleaned.csv')
```
