```python
import pdfplumber
import pandas as pd
```

```python

pdf_file_path = r'D:\nc-2023_q2_citssregistrantsreport.pdf'


all_tables = []

with pdfplumber.open(pdf_file_path) as pdf:
    # Iterate through the pages
    for page in pdf.pages:
        # Extract tables from the page
        tables = page.extract_tables()

        # Append the tables to the list
        all_tables.extend(tables)

# Flatten the list of lists
flattened_tables = [item for sublist in all_tables for item in sublist]

# Convert the flattened list into a single DataFrame
combined_dataframe = pd.DataFrame(flattened_tables)

# Print or use the combined DataFrame
print(combined_dataframe)
```
