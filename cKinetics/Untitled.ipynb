```python
import datetime
import pandas as pd
import requests
from datetime import datetime as dt

#338 ethanol futures
#8381 german power
#312 soyabean oil

def CME(access_token, table_name, code):
    base_url = "https://www.cmegroup.com/CmeWS/mvc/Settlements/Futures/Settlements"
    end_date = datetime.date.today()  # Get the current date
    start_date = end_date - datetime.timedelta(days=7)  # Calculate the start date

    data_frames = []  # List to store DataFrames for each day

    while start_date <= end_date:
        date_str = start_date.strftime("%m/%d/%Y")  # Format date as "MM/DD/YYYY"
        url = f"{base_url}/{code}/FUT"
        params = {
            "strategy": "DEFAULT",
            "tradeDate": date_str,
            "pageSize": "500"
        }

        headers = {
            "accept": "application/json, text/plain, */*"
        }

        response = requests.get(url, params=params, headers=headers)

        data = response.json()
        settlements = data.get("settlements", [])

        df = pd.DataFrame(settlements)
        df['date'] = start_date.strftime("%d-%m-%Y")  # Format date as "DD-MM-YYYY"
        data_frames.append(df)

        start_date += datetime.timedelta(days=1)  # Move to the next day

    # Concatenate all DataFrames into a single DataFrame
    result_df = pd.concat(data_frames, ignore_index=True)

    print(result_df)
    #push_data2(commit_table[table_name], result_df, access_token)
    
    #result_df.to_csv(r'D:\soyabean_oil.csv')
```
