```python
import requests

def get_first_email(access_token, sender_email, subject_substring):
    # Define the Microsoft Graph API endpoint for messages
    messages_endpoint = "https://graph.microsoft.com/v1.0/me/messages"

    # Define the query parameters to filter messages based on the sender email and subject substring
    query_params = {
        "$filter": f"from/emailAddress/address eq '{sender_email}' and contains(subject, '{subject_substring}')"
    }

    # Set up the request headers, including the access token
    headers = {
        "Authorization": f"Bearer {access_token}",
        "Content-Type": "application/json"
    }

    # Make the request to the Microsoft Graph API
    response = requests.get(messages_endpoint, headers=headers, params=query_params)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the response JSON
        messages = response.json().get("value", [])

        # Check if there are any messages
        if messages:
            # Get the first message
            first_message = messages[0]
            print(first_message['createdDateTime'])

            print(f"Subject: {first_message['subject']}")
            
            # Check if the message has a text/plain body
            if 'text' in first_message['body']:
                print(f"Text Body: {first_message['body']['text']}")
            else:
                # If not, print the HTML body as text
                print(f"HTML Body (as text): {first_message['body']['content']}")
            
            print("-" * 50)
        else:
            print("No emails found.")
    else:
        # Print an error message if the request was not successful
        print(f"Error: {response.status_code} - {response.text}")

# Replace 'YOUR_ACCESS_TOKEN' with your actual access token
#access_token = 'YOUR_ACCESS_TOKEN'

# Replace 'o.akcasoy@greenipath.com' with the sender email you're interested in
sender_email = 'o.akcasoy@greenipath.com'

# Replace 'Greenipath New Available IREC/TIGR' with the subject substring you're looking for
subject_substring = 'Greenipath New Available IREC/TIGR'

# Call the function to get and print the first email
get_first_email(access_token, sender_email, subject_substring)
```

```python
import requests
import html2text

def get_nth_email(access_token, sender_email, subject_substring, n):
    # Define the Microsoft Graph API endpoint for messages
    messages_endpoint = "https://graph.microsoft.com/v1.0/me/messages"

    # Define the query parameters to filter messages based on the sender email and subject substring
    query_params = {
        "$filter": f"from/emailAddress/address eq '{sender_email}' and contains(subject, '{subject_substring}')"
    }

    # Set up the request headers, including the access token
    headers = {
        "Authorization": f"Bearer {access_token}",
        "Content-Type": "application/json"
    }

    # Make the request to the Microsoft Graph API
    response = requests.get(messages_endpoint, headers=headers, params=query_params)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the response JSON
        messages = response.json().get("value", [])

        # Check if there are any messages
        if messages and len(messages) > n:
            # Get the nth message (in this case, the 5th message)
            nth_message = messages[n - 1]
            date_received = nth_message['createdDateTime']
            print(f"Date Received: {date_received}")
            print(f"Subject: {nth_message['subject']}")
            
            # Check if the message has a text/plain body
            if 'text' in nth_message['body']:
                print(f"Text Body: {nth_message['body']['text']}")
            else:
                # If not, convert the HTML body to plain text using html2text
                html_body = nth_message['body']['content']
                text_body = html2text.html2text(html_body)
                print(f"HTML Body (as text): {text_body}")
            
            print("-" * 50)
        else:
            print(f"No {n}th email found.")
    else:
        # Print an error message if the request was not successful
        print(f"Error: {response.status_code} - {response.text}")

# Replace 'YOUR_ACCESS_TOKEN' with your actual access token
# access_token = 'YOUR_ACCESS_TOKEN'

# Replace 'o.akcasoy@greenipath.com' with the sender email you're interested in
sender_email = 'o.akcasoy@greenipath.com'

# Replace 'Greenipath New Available IREC/TIGR' with the subject substring you're looking for
subject_substring = 'Available IREC/TIGR'

# Call the function to get and print the 5th email
get_nth_email(access_token, sender_email, subject_substring, n=3)
```
