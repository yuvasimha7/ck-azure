```python
driver = webdriver.Chrome()

driver.get("https://commodityinsights.spglobal.com/GlobalCarbonMarkets2023.html")
#webdriver.wait(driver,15).until(EC.presence_of_element_located(("xpath", '/html/body/div[1]/header[1]/div/div[1]/a')))
'''element_present = EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div[3]/main/div/div[2]/div/div/article/div/div/div/div/div/ul/li[1]/div/div[1]/h3/a'))

# Wait for a maximum of 15 seconds before throwing an exception
element = WebDriverWait(driver, 15).until(element_present)'''
name = []
company=[]

for i in range(2, 54):
    xpath1 = f"/html/body/div[4]/div/div/div[{i}]/div/span/span[1]"
    xpath2 = f"/html/body/div[4]/div/div/div[{i}]/div/span/span[1]/span"
    
    # Find the element using XPath
    element1 = driver.find_element(By.XPATH, xpath1)
    element2 = driver.find_element(By.XPATH, xpath2)
    #print(xpath1)
    #print(element1)
    print(element1.text)
    # Get the text and append it to the list
  
    name.append(element1.text)
    company.append(element2.text)
# Print or use the extracted texts as needed
#data = {'Name': name, 'Company': company}
data = {'Name': name}
df = pd.DataFrame(data)

# Close the WebDriver
driver.quit()
```
